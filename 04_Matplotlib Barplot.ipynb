## Matplotlib Barplot
    A bar plot (bar chart) in Matplotlib is used to represent categorical data with rectangular bars, where the height (or length) of each bar corresponds to a value. It is one of the most commonly used visualizations for comparing quantities across different categories.

    -> A bar plot displays:

     - Categories → on the X-axis (e.g., languages, products, students)
     - Values → on the Y-axis (e.g., counts, marks, sales)

    🔹 Key Customization Options
    
      - color → Change bar color
      - width → Adjust bar thickness
      - label → Add legend
      - edgecolor → Border styling
      - alpha → Transparency

```python
# Import required libraries
import pandas as pd
import numpy as np
import random 
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
```

###  Generating Sample Data For plotting

```python
# Define categories (programming languages)
Classes = ["AI", "ML", "Js", "DS", "Python"]

# Number of students in each class
Class_1 = [60, 98, 67, 59, 80]
Class_2 = [76, 66, 34, 55, 61]
Class_3 = [88, 61, 59, 42, 23]
```

### EX 1 - Single Class_1 Students Barplot

```python
# Create a bar chart for Class 1 students
plt.bar(
    Classes, Class_1,
    width=0.4,                 # Width of bars
    align='center',            # Align bars at center
    edgecolor='black',         # Border color
    color="y",                 # Fill color (yellow)
    linewidth=1,               # Border thickness
    alpha=0.9,                 # Transparency level
    label="Class_1 Students"   # Legend label
)

# Define font styles for title and labels
font1 = {'family':'serif','color':'#473505','size':18}
font2 = {'family':'serif','color':'#a3044E','size':15}


# Add title and axis labels
plt.title("Programming Language Students", fontdict = font1)
plt.xlabel("Languages",fontdict= font2)
plt.ylabel("No. of Students",fontdict  = font2)
plt.legend()
plt.show()
```

### Multiple Class Students Barplots

```python
# Create an array of indices based on the number of Classes (categories)
classes_index = np.arange(len(Classes))  
index = np.arange(len(Classes))
width = 0.2 # Define the width of each bar

# Plot bar chart for Class 1 students
# Bars are placed at the original index positions
plt.bar(classes_index, Class_1, width,color="#0A8162D0", label="Class 1 Students")  

# Plot bar chart for Class 2 students
# Bars are shifted to the right by 'width' to avoid overlap
plt.bar(classes_index + width, Class_2, width, color="#C78C1ECA", label="Class 2 Students")

# Plot bar chart for Class 3 students
# Bars are shifted further right by '2 * width'
plt.bar(classes_index + width + width, Class_3, width,color="#CD4D1A", label="Class 3 Students")

# Set the title and label of the bar chart
plt.title("Programming Lang. students", fontdict=font1)
plt.xlabel("Languages", fontdict=font2)
plt.ylabel("No. of Students", fontdict=font2)
plt.xticks(index + width,Classes)
plt.legend()  # Display legend to identify each class
plt.show()
```

### EX 2 - DAILY TRADING VOLUME

```python
stocks = ['HDFC', 'TATASILV', 'MAZDOCK', 'ADANIENS', 'NIPPON']  # Stock names
volume = [1151.5, 950, 1500, 1198, 800] # Stock Volume ( in thousands )

# Create bar plot
plt.bar(stocks, volume, color='#085A4A')
plt.title('Daily Trading Volume of Stocks')
plt.xlabel('Stock Name')
plt.ylabel('Volume ( in thousands )')
plt.show()
```

### EX 3 -COMPNIES STOCK CAMPARISION

```python
stocks = ['HDFC', 'TATASILV', 'MAZDOCK','ADANIES', 'NIPPON']  # Stock names
 
# Stock prices
open_price = [3400, 1600, 2400,2800,2650]
close_price = [3450, 1570, 2900,3000,2600]
high_price = [3480, 2200, 3520,3100,3690]

# Create index for stocks
group_gap = 0.5
index = np.arange(len(stocks)) * (3 * 0.25 + group_gap)

# Width of each bar
width = 0.2

# Plot multiple bars side by side
plt.bar(index, open_price, width, label='Open Price')
plt.bar(index + width, close_price, width, label='Close Price')
plt.bar(index + 2*width, high_price, width, label='High Price')

# Labels and title
plt.xlabel('Stocks')
plt.ylabel('Price')
plt.title('Stock Price Comparison')
plt.xticks(index + width, stocks)
plt.legend()
plt.show()
```
