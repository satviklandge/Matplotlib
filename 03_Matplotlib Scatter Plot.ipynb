## Matplotlib Scatterplot 
    A scatter plot in Matplotlib is a fundamental visualization used to display the relationship between two numerical variables. Each point on the graph represents one observation from the dataset, plotted using its (x, y) coordinates.

        Basic Purpose ::
    - Show how one variable changes with another
    - Identify patterns, trends, clusters, or outliers

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load full dataset
df = pd.read_csv('googleplaystore.csv')
df
```

```python
# Load only first 1000 rows (for faster processing)
df1 = pd.read_csv('googleplaystore.csv', nrows=1000)
df1.head(5)
```

```python
# Show column names
df1.columns
```

```python
# Check shape (rows, columns)
data = pd.read_csv("googleplaystore.csv",nrows=1000)
data.shape
print("No. of rows :",data.shape[0])
print("No. of cols :",data.shape[1])
```

```python
# Define font style for title
font1 = {
    'family': 'serif',
    'color': '#840773',
    'size': 18
}
font2 = {
    'family': 'serif',
    'color': "#D07418",
    'size': 15
}
```

```python
# SINGLE SCATTER PLOT
# -----------------------------

# Extract columns for plotting

x = df1['Rating']
y = df1['Reviews']

# Create scatter plot
plt.scatter(x, y, marker='*', edgecolor='black', alpha=0.6, color='yellow')

# Set the title and label of the scatter plot
plt.title("Google Play Store Apps Scatter plot", fontdict=font1)
plt.xlabel("Rating", fontdict=font2)
plt.ylabel("Reviews", fontdict=font2)
plt.show()
```

## Example - Google Play Store Apps: Rating vs Reviews & Installs

```python
# MULTIPLE SCATTER PLOTS
# -----------------------------

x = data['Rating']    # X-axis values: App ratings
y = data['Reviews']   # Y-axis values for first scatter plot: Number of reviews

# Scatter plot for Reviews
plt.scatter(x, y, marker='*', edgecolor='#8C4B0B',alpha=0.5,color='#8C4B0B', label='Reviews')

# Scatter plot for Installs (same X-axis, different Y-axis)
plt.scatter(x, data['Installs'],marker='o',edgecolor='#08785E',alpha=0.5,color='#08785E', label='Installs')

# Set the title and label of the plot
plt.title("Google Play Store Apps Scatter plot", fontdict=font1)
plt.xlabel("Rating", fontdict=font2)
plt.ylabel("Reviews & Installs", fontdict=font2)
plt.legend(loc=2)
plt.show()
```
