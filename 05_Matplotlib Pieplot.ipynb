## Matplotlib Pie Chart 
    A Pie Chart in Matplotlib is a circular statistical graphic used to represent data as proportions of a whole. Each slice (or wedge) corresponds to a category, and its size reflects the relative contribution of that category to the total.

    -> Important Parameters
        
    | Parameter    | Description                    |
    | ------------ | ------------------------------ |
    | `x`          | Data values (required)         |
    | `labels`     | Names of each category         |
    | `colors`     | Custom colors for slices       |
    | `autopct`    | Displays percentage values     |
    | `explode`    | Separates a slice for emphasis |
    | `shadow`     | Adds shadow effect             |
    | `startangle` | Rotates the pie chart          |


```python
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use("ggplot")
```

## Basic Pie Chart

```python
# BASIC PIE CHART (SIMPLE EXAMPLE)
# --------------------------------------------------

# Plot a very simple pie chart with two values
plt.pie([1,2])
plt.show
```

## Ex 1 - Marks of Student in each Class 

```python
# PIE CHART WITH CUSTOM LABELS AND COLORS
# --------------------------------------------------

# Categories (labels)
Classes = ['Machine Learning', 'Artificial Intelligence', 'Python', 'JavaScript', 'Data Science']

# Number of students in each class
Class1_Students = [67, 98, 54, 76, 55]

# Custom colors for each slice
colors = ['red', 'lightblue', 'green', 'orange', 'purple']

# Plot pie chart
plt.pie(Class1_Students,
        labels = Classes,
        colors = colors)
plt.show()
```

```python
# PIE CHART WITH EXPLODE + TEXT CUSTOMIZATION
# --------------------------------------------------

# Explode: separates one slice from the pie (highlight)
explode = [0.2,0,0,0,0]
colors = ['red', 'lightblue', 'green', 'orange', 'purple']


# Text properties (font size)
textprops = {"fontsize":15}  


# Plot pie chart with customization
plt.pie(
    Class1_Students,
    labels=Classes,
    explode=explode,              # Highlight first slice
    colors=colors,
    textprops=textprops,          # Customize text size
    autopct='%0.1f%%',            # Show percentage with 1 decimal
    shadow=True,                  # Add shadow effect
    startangle=90                 # Rotate chart for better alignment
)
plt.show()
```

## Ex 2 - Purchase Price Of Electronics

```python
# REAL-WORLD EXAMPLE (PRODUCT PRICES)
# --------------------------------------------------

# Item names
Items_name = ['TV', 'AC', 'Freeze', 'Mobile', 'Cooler']

# Prices of items
Price = [15000, 20000, 17000, 18500, 20500]

# Highlight second item (AC)
explode = (0, 0.2, 0, 0, 0)

# Plot pie chart
plt.pie(
    Price,
    labels=Items_name,
    explode=explode,             # Highlight AC
    autopct='%1.3f%%',           # Show percentage (3 decimal places)
    startangle=120               # Rotate chart
)

plt.title("Price Distribution of Electronic Items")
plt.show()
```
