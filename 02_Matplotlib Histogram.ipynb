##  Matplotlib Histogram
    A histogram in Matplotlib is a graphical representation used to visualize the distribution of numerical data.
    It groups data into intervals called bins and shows how many values fall into each interval.
    It form a rectangle bar,each bar shows some data, which belong to different categories. 

        -> Key Parameters
        
           - data → Input numerical dataset
           - bins → Number or range of intervals
           - color → Bar color
           - rwidth → Width of bars (spacing control)
           - histtype → Type of histogram (bar, step, etc.)
           - orientation → Vertical or horizontal
            
                

```python
# IMPORT REQUIRED LIBRARIES
import matplotlib.pyplot as plt
import numpy as np 
import random 
```

```python
# GENERATE SAMPLE DATA
# --------------------------------------------------

# Generate random ages for Machine Learning students (100 students, age 18–45)
ml_students_age = np.random.randint(18, 45, (100))

# Generate random ages for Python students (100 students, age 15–40)
py_students_age = np.random.randint(15, 40, (100))
```

### Display Ml Student Ages

```python
ml_students_age
```

### Display Py Student Ages

```python
py_students_age
```

```python
# BASIC HISTOGRAM (ML STUDENTS ONLY)
# --------------------------------------------------

# Plot histogram for ML student ages
plt.hist(ml_students_age)

# Add labels and title
plt.xlabel('Student Age Category')
plt.ylabel('No. of Students')
plt.title('ML Student Age Record')
```

```python
# CUSTOMIZED HISTOGRAM (ML STUDENTS)
# --------------------------------------------------

# Set figure size
plt.figure(figsize=(16, 9))

# Define custom bins (age groups)
bins = [15, 20, 25, 30, 35, 40, 45]

# Plot histogram with customization
plt.hist(
    ml_students_age,
    bins=bins,
    rwidth=0.8,              # Bar width
    histtype='bar',          # Histogram type
    color='y',               # Color: yellow
    label='ML Student Age',
    orientation='vertical'
)

# Show legend
plt.legend()
```

```python
# COMPARISON HISTOGRAM (ML vs PYTHON STUDENTS)
# --------------------------------------------------

# Set figure size
plt.figure(figsize=(16, 9))

# Define bins again
bins = [15, 20, 25, 30, 35, 40, 45]

# Plot combined histogram
plt.hist(
    [ml_students_age, py_students_age],
    rwidth=0.8,
    histtype="bar",
    color=['gold', 'lightgreen'],
    label=["ML Student", "Py Student"]
)

# Add title and labels
plt.title("ML & Py Student Age Histograms")
plt.xlabel("Student Age Category")
plt.ylabel("No. of Students")

# Show legend
plt.legend()

# Display final plot
plt.show()
```
