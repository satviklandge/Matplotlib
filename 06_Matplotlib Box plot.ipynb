## Matplotlib Boxplot
    A boxplot (also called a box-and-whisker plot) in Matplotlib is used to visualize the distribution of numerical data through key statistical measures.

        -> What a Boxplot Shows
    
         A boxplot summarizes data using:   
         
        - Minimum – smallest value (excluding outliers)
        - Q1 (25th percentile) – lower quartile
        - Median (Q2) – middle value
        - Q3 (75th percentile) – upper quartile
        - Maximum – largest value (excluding outliers)
        - Outliers – values that fall far outside the normal range
        

            ->  Important Parameters
            
            - patch_artist=True → fills the box with color
            - notch=True → shows confidence interval around median
            - vert=False → makes horizontal boxplot
            - showmeans=True → displays mean
            - labels → adds category names (for multiple datasets)

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.style as style 
style.use("ggplot")
```

## Basic Box plot 

```python
# Sample Data
data = [10, 20, 30, 40, 50, 100]

plt.boxplot(data)
plt.title("Simple Boxplot")
plt.show()
```

## Example 1 - Boxplot vs Histogram

```python
# Generate data
np.random.seed(0)
data = np.random.normal(50, 10, 200)

plt.figure(figsize=(12,5))

# Histogram
plt.subplot(1,2,1)
plt.hist(data, bins=20)
plt.title("Histogram (Distribution)")
plt.xlabel("Values")
plt.ylabel("Frequency")

# Boxplot
plt.subplot(1,2,2)
plt.boxplot(data)
plt.title("Boxplot (Summary)")

plt.tight_layout()
plt.show()
```

## Example 2 -  Student performance Analysis By Boxplot

```python

# Simulated student scores
np.random.seed(10)
class_A = np.random.normal(70, 10, 100)
class_B = np.random.normal(65, 12, 100)
class_C = np.random.normal(75, 8, 100)
class_D = np.random.normal(60, 15, 100)

data = [class_A, class_B, class_C, class_D]

plt.figure(figsize=(12, 6))

plt.boxplot(
    data,
    patch_artist=True,
    labels=["Class A", "Class B", "Class C", "Class D"],
    showmeans=True
)

# Custom colors
colors = ["skyblue", "lightgreen", "orange", "pink"]
for patch, color in zip(plt.boxplot(data, patch_artist=True)['boxes'], colors):
    patch.set_facecolor(color)

plt.title("Student Score Distribution by Class")
plt.ylabel("Scores")
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```
